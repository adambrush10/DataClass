```python
file1 ='/Users/Brush/PycharmProjects/python-challenge/PyCitySchools/Homeworks_HW04-Pandas_PyCitySchools_Resources_schools_complete.csv'
file2 = '/Users/Brush/PycharmProjects/python-challenge/PyCitySchools/Homeworks_HW04-Pandas_PyCitySchools_Resources_students_complete.csv'

schools_df = pd.read_csv(file1)
students_df = pd.read_csv(file2)
```

```python
# Merge both df's with the school name as the index

merge_table = pd.merge(schools_df, students_df, on="school_name")
merge_table.head()
```

```python
#Group by type (District and Charter)
District_type = merge_table.loc[merge_table["type"] == "District", :]

charter_type = merge_table.loc[merge_table["type"] == "Charter", :]
```

```python
   # TYPE SUMMARY

#create a key metrics table by Types

grouped_type = merge_table.groupby(['type'])

#number of schools
total_schools = grouped_type['school_name'].nunique()

#number of students
student_count = grouped_type['Student_ID'].count()

#total budget
budget_sum = grouped_type['budget'].sum()

#average Math score
avg_math = grouped_type['math_score'].mean()

#average reading score
avg_reading = grouped_type['reading_score'].mean()

# % passing math

m_pass = 0
m_fail = 0

for grades in charter_type['math_score']:
    if grades >= 70:
        m_pass += 1
    else:
        m_fail += 1


charter_math_pass_pct = m_pass / 12194
math_fail_pct = m_fail / 12194


m2_pass = 0
m2_fail = 0

for grades in District_type['math_score']:
    if grades >= 70:
        m2_pass += 1
    else:
        m2_fail += 1


District_math_pass_pct = m2_pass / 26976


# % passing reading

r_pass = 0
r_fail = 0

for grades in charter_type['reading_score']:
    if grades >= 70:
        r_pass += 1
    else:
        r_fail += 1


charter_reading_pass_pct = r_pass / 12194
reading_fail_pct = r_fail / 12194


r2_pass = 0
r2_fail = 0

for grades in District_type['reading_score']:
    if grades >= 70:
        r2_pass += 1
    else:
        r2_fail += 1


District_reading_pass_pct = r2_pass / 26976
```

```python

#overall_passing_rate

District_overall_pass_pct = (District_math_pass_pct + District_reading_pass_pct) / 2
charter_overall_pass_pct = (charter_math_pass_pct + charter_reading_pass_pct) / 2


#master table for Type

s1 = total_schools
s2 = student_count
s3 = budget_sum
s4 = avg_math
s5 = avg_reading


master_table = pd.concat([s1, s2, s3, s4, s5], axis=1)

master_table['math_passing_percentage'] = [charter_math_pass_pct, District_math_pass_pct]
master_table['reading_passing_percentage'] = [charter_reading_pass_pct, District_reading_pass_pct]
master_table['overall_passing_percentage'] = [District_overall_pass_pct, charter_overall_pass_pct]

print('Type Summary')
master_table
```
