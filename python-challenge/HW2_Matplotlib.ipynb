```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
city_data = '/Users/Brush/UCIRV201810DATA4/Homeworks/HW05-Matplotlib/Pyber/data/city_data.csv'
ride_data = '/Users/Brush/UCIRV201810DATA4/Homeworks/HW05-Matplotlib/Pyber/data/ride_data.csv'

```

```python
city_df = pd.read_csv(city_data)
ride_df = pd.read_csv(ride_data)

#merge both df's into one 

city_ride_df = pd.merge(city_df, ride_df, on='city')
city_ride_df
```

```python
city_ride_df.count()
```

```python
#calculate average fare by city

avg_fare_by_city = city_ride_df[['fare', 'city']].groupby(['city']).mean()
avg_fare_by_city.head()
```

```python
#calculate total number of riders per city 

total_riders_by_city = city_ride_df[['ride_id', 'city']].groupby(['city']).count()
total_riders_by_city.head()
```

```python
#total number of drivers per city 

total_drivers_by_city = city_ride_df[['driver_count', 'city']].groupby(['city']).sum()
total_drivers_by_city.head()
```

```python
# City grouped by city type
type_by_city = city_ride_df[['type', 'city']].groupby(['type']).count()
type_by_city.head()
```

```python
city_ride_df = city_ride_df.merge(avg_fare_by_city, left_on='city', right_on='city', how='outer')
city_ride_df = city_ride_df.merge(total_riders_by_city, left_on='city', right_on='city', how='outer')
city_ride_df = city_ride_df.merge(total_drivers_by_city, left_on='city', right_on='city', how='outer')

master_df = city_ride_df.rename(columns={'fare_y': 'Average Fare', 'ride_id_y': 'Total Riders in the City', 'driver_count_y': 'Driver Count in City'})
#master_df.head()

```

```python
x = total_riders_by_city
y = avg_fare_by_city
z = total_drivers_by_city
c = type_by_city
 
# use the scatter function
plt.scatter(x, y, s=z, Alpha=.5)

# Change color with c and alpha. I map the color to the X axis value.
#plt.scatter(c, cmap="Blues", edgecolors="grey", linewidth=2)
 
# Add titles (main and on axis)
plt.xlabel("total riders")
plt.ylabel("Average Fare")
plt.title("Pyber Ride Sharing Data")

plt.show()
```

```python
# % of Total Fares by City Type

type_by_city = city_ride_df[['type', 'city']].groupby(['type']).count()


total_cities = type_by_city.sum()
type_by_city.head()
```

```python
rural = type_by_city.iloc[0, 0]
print(rural)
suburban = type_by_city.iloc[1, 0]
print(suburban)
urban = type_by_city.iloc[2, 0]
print(urban)
```

```python
rural_percentage = rural / total_cities
print(rural_percentage)
suburban_percentage =  suburban / total_cities
print(suburban_percentage)
urban_percentage =  urban / total_cities
print(urban_percentage)
```

```python
# % of Total Fares by City Type -- Pie Chart


total_fare_by_city = city_ride_df[['fare', 'type']].groupby(['type']).sum()


rural_fare = total_fare_by_city.iloc[0, 0]
print(rural_fare)
suburban_fare = total_fare_by_city.iloc[1, 0]
print(suburban_fare)
urban_fare = total_fare_by_city.iloc[2, 0]
print(urban_fare)

total_fare_cities = total_fare_by_city.sum()

rural_fare_percentage = rural_fare / total_fare_cities
print(rural_fare_percentage)
suburban_fare_percentage =  suburban_fare / total_fare_cities
print(suburban_fare_percentage)
urban_fare_percentage =  urban_fare / total_fare_cities
print(urban_fare_percentage)




total_fare_by_city.head()
```

```python
# % of Total Fare by City Type

labels = 'Rural', 'Suburban', 'Urban'
sizes = [rural_fare_percentage, suburban_fare_percentage, urban_fare_percentage]
colors = ['lightblue', 'red', 'orange']
explode = (0.1, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```
